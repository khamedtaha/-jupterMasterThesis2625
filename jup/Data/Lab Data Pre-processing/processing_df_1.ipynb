## Feauter fees

```python
print(df_1['fees'].isna().sum())
```

```python
df_1['fees'] = df_1['fees'].str.extract(r'(\d+\.?\d*)')

df_1['fees'] = pd.to_numeric(df_1['fees'], errors='coerce')

df_1['fees'] = df_1['fees'].fillna(0.0)

print(df_1['fees'].dtype)
```

```python
df_1.info()
```

## Feauter specialization

```python
print(df_1['specialization'].isna().sum())
```

```python
df_1['specialization'] = df_1['specialization'].fillna('unknown')

print(f"NaNs in specialization: {df_1['specialization'].isna().sum()}")
```

## Feauter waiting_time

```python
# df_1.head(100)
```

```python
print(df_1['waiting_time'].isna().sum())
```

```python
df_1['waiting_time_pre'] = df_1['waiting_time']
```

```python
df_1.head()
```

```python
import pandas as pd

df_1['waiting_time_pre'] = df_1['waiting_time_pre'].fillna("Waiting Time : 0 Minutes")

# SyntaxWarning
hours = df_1['waiting_time_pre'].str.extract(r'(\d+)\s*Hour').fillna(0).astype(int)

minutes = df_1['waiting_time_pre'].str.extract(r'(\d+)\s*Minute').fillna(0).astype(int)

df_1['waiting_time_minutes'] = (hours[0] * 60) + minutes[0]


print(df_1[['waiting_time_pre', 'waiting_time_minutes']].head(10))
```

```python
df_1.info()
```

```python
df_1 = df_1.drop(columns=['waiting_time_pre'])
```

```python
df_1.info()
```

```python
# Detailed visualization for Doctor Fees
plt.figure(figsize=(12, 5))

# Histogram
plt.subplot(1, 2, 1)
sns.histplot(df_1['fees'], kde=True, color='green')
plt.title('Distribution of Fees (EGP)')

# Boxplot to see outliers
plt.subplot(1, 2, 2)
sns.boxplot(x=df_1['fees'], color='green')
plt.title('Boxplot of Fees')

plt.tight_layout()
plt.show()
```

```python
# Detailed visualization for Ratings
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(df_1['avg_rate'], kde=True, color='orange', bins=10)
plt.title('Distribution of Average Ratings')

plt.subplot(1, 2, 2)
sns.violinplot(x=df_1['avg_rate'], color='orange')
plt.title('Density of Ratings')

plt.tight_layout()
plt.show()
```

```python
# Detailed visualization for Views
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(df_1['doctor_views'], kde=True, color='blue')
plt.title('Distribution of Profile Views')

plt.subplot(1, 2, 2)
sns.boxplot(x=df_1['doctor_views'], color='blue')
plt.title('Identification of Most Viewed Doctors')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the visual style for better clarity
sns.set_theme(style="whitegrid")

# Create a figure with two side-by-side plots
plt.figure(figsize=(12, 5))

# 1. Histogram with KDE (To see the shape of the distribution)
plt.subplot(1, 2, 1)
sns.histplot(df_1['waiting_time_minutes'], kde=True, color='red', bins=20)
plt.title('Distribution of Waiting Time (Min)', fontsize=14)
plt.xlabel('Minutes')
plt.ylabel('Frequency')

# 2. Boxplot (To identify doctors with extreme waiting times/outliers)
plt.subplot(1, 2, 2)
sns.boxplot(x=df_1['waiting_time_minutes'], color='red')
plt.title('Waiting Time Outliers', fontsize=14)
plt.xlabel('Minutes')

# Adjust layout to prevent title overlapping
plt.tight_layout()
plt.show()
```
