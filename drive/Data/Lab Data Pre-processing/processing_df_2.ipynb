```python
import re
def clean_text(text):
    if not isinstance(text, str):
        return ""
    # Remove [XXXX] placeholders
    text = re.sub(r'XXXX', '', text)
    # Remove multiple spaces
    text = re.sub(r'\s+', ' ', text).strip()
    return text
```

```python
df_2['Patient_Query'] = df_2['Patient'].apply(clean_text)
df_2['problem_description'] = df_2['problem_description'].apply(clean_text)
```

```python
# df_2.head()
```

```python
df_2['Doctor_Response'] = df_2['Doctor'].str.extract(r'<start>(.*?)<end>', flags=re.DOTALL, expand=False).str.strip()
```

```python
df_2 = df_2.rename(columns={
    'PD_ID': 'ID',
    'Description': 'question_title',
    'Doctor_Response': 'doctor_response',
    'Patient_Query'  : 'patient_query',
    'problem_description': 'problem_content',
})
```

```python
selected_columns = ['ID', 'question_title', 'patient_query', 'problem_content','doctor_response']

df_2_new = df_2[selected_columns].copy()
```
