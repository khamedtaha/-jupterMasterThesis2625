```python
import io

def get_dfs_summary(df_dict):
    output = io.StringIO()
    for name, df in df_dict.items():
        output.write(f"=== Summary for {name} ===\n")
        output.write(f"Columns: {df.columns.tolist()}\n")
        output.write(f"Shape: {df.shape}\n")

        buffer = io.StringIO()
        df.info(buf=buffer)
        output.write(f"Info:\n{buffer.getvalue()}\n")

        output.write(f"Sample Data:\n{df.head(2).to_string()}\n")
        output.write("-" * 50 + "\n\n")

    return output.getvalue()

my_dataframes = {
    "df_1": df_1,
    "df_2": df_2,
    "df_4": df_4,
    "df_5": df_5
}

final_text = get_dfs_summary(my_dataframes)
print(final_text)
```
